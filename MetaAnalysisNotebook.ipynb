# Analysis

## Meta-analysis

Assessing the relationship between ADHD symptomology and caffeinated beverage consumption

```r
new_res = rma(yi, vi, data=new_efz)
new_res
predict(new_res, digits=3, transf=transf.ztor)
confint(new_res)
```

#### Interpretation

The random-effects model gave a significant pooled Fisher's Z effect size of 0.1141 with a p-value of \<.0001. These results show a weak and positive association between ADHD symptoms and energy drink consumption.

An I\^2 value of 92.07% indicates a large amount of heterogeneity between studies.

## Influence Test

Run an influence test to see if there were any studies that were particularly influential - if results hang on a particular study (data point is red), that means that study had a significant influence

```r
inf = influence(new_res)
print(inf)
plot(inf)
```

### Interpretation

No studies were plotted in red. This indicates that no studies had a significant influence.

## Funnel plot

Assessing publication bias

```r
estimate = 0.1141
se = 0.0163

#Store a vector of values that spans the range from 0
#to the max value of impression (standard error) in your dataset.
#Make the increment (the final value) small enough (I choose 0.001)
#to ensure your whole range of data is captured
se.seq=seq(0, max(new_efz$se), 0.001)

#Compute vectors of the lower-limit and upper limit values for
#the 95% CI region
ll95 = estimate-(1.96*se.seq)
ul95 = estimate+(1.96*se.seq)

#Do this for a 99% CI region too
ll99 = estimate-(3.29*se.seq)
ul99 = estimate+(3.29*se.seq)

#Calculate the confidence interval for your meta-analytic estimate
meanll95 = estimate-(1.96*se)
meanul95 = estimate+(1.96*se)

#Put all calculated values into one data frame
dfCI = data.frame(ll95, ul95, ll99, ul99, se.seq, estimate, meanll95, meanul95)


#Draw Plot
library(ggplot2)
fp = ggplot(aes(x = se, y = yi), data = new_efz) +
    geom_point(shape = 1)+
    xlab('Standard Error') + ylab('Effect Size')+
    geom_line(aes(x = se.seq, y = ll95), linetype = 'dotted', data = dfCI) +
    geom_line(aes(x = se.seq, y = ul95), linetype = 'dotted', data = dfCI) +
    geom_line(aes(x = se.seq, y = ll99), linetype = 'dashed', data = dfCI) +
    geom_line(aes(x = se.seq, y = ul99), linetype = 'dashed', data = dfCI) +
    geom_segment(aes(x = min(se.seq), y = meanll95, xend = max(se.seq), yend = meanll95), linetype='dotted', data=dfCI) +
    geom_segment(aes(x = min(se.seq), y = meanul95, xend = max(se.seq), yend = meanul95), linetype='dotted', data=dfCI) +
    scale_x_reverse()+
    scale_y_continuous(breaks=seq(-1.25,2,0.25))+
    coord_flip()+
    theme_bw()
fp
```

#### Interpretation

In a funnel plot, larger studies, which have a smaller standard error, are plotted high on the y-axis, while smaller studies, which have a larger standard error, are plotted lower on the y-axis.

After viewing the funnel plot, it appears their is potential asymmetry, indicating potential study bias. We can see that effect sizes are smaller at the very top of the funnel. This indicates that studies with smaller sample sizes tending to reporting larger effect sizes and may be biasing the overall effect size. This could indicate that smaller studies with smaller effect sizes are not being reported.

## Meta-regression

Assessing potential moderators in the relationship between ADHD symptomatology and caffeinated beverage consumption

### Moderating Effect of Age

```r
new_res.modage = rma(yi, vi, mods = ~ age, data=new_efz)
new_res.modage
```

#### Interpretation

The age coefficient of the model is 0.0003 (p=0.8673). This estimate represents the change in the effect size for each one-unit increase in the moderator variable and is not statistically significant. This indicates participant age did not have a significant moderating effect on the relationship between ADHD symptomology and caffeinated beverage consumption.

#### Visualizing the moderating relationship

with a regression plot (aka bubble plot)

Studies with larger weights have larger 'bubbles' on the plot

Larger studies with smaller standard errors have higher weights in the model

```r
regplot(new_res.modage, xlab='Age as a Moderator', ylab='Effect Size',grid = TRUE)
```

##### Interpretation

We can see that in the regression plot, there is little to no moderating effect of age on the model - slope is nearly zero.

## Robust variance estimation

accounting for effect size dependencies

Some of the effect sizes came from the same study

```r
mes_ss = robu(formula = yi ~ 1, data = new_efz, studynum = authors,
              var.eff.size = vi, small = TRUE)
mes_ss
```

### Interpretation

After accounting for effect size dependencies, the effect size remained statistically significant (p-value 0.000506).
